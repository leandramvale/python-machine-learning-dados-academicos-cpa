#Limpeza dos Dados e Configuração do Dataset

Observa-se que temos valores não informados em alguns indicadores-respostas. Isso se deve ao fato de que pode ser que o professor não tenha recebido nenhuma resposta para este indicador em todas as disciplinas que ele ministra aulas.

````python
#assim sendo vamos colocar o valor zero para os valores ausentes na frequencia das respostas dadas.
df.fillna(0, inplace=True)
df.head()
````

Observa-se também que temos tipos de dados float64 e int64, e portanto seria interessante converter tudo para float64, deixando apenas o codprof como int64.

````python
#todos os tipos de dados que são int64 devem ser convertidos para float64 exceto a coluna codprof
for col in df.columns:
    if col != 'CODPROF':
        df[col] = df[col].astype('float64')
````
